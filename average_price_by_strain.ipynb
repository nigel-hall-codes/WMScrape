```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

conn = sqlite3.connect("Medigreen.db")
cursor = conn.cursor()
```

```python
q = """
    select product.name, avg(price.price) from producthasprice as php
    join product on product.id=php.productid
    join price on price.id=php.priceid
    where price.price > 0 and price.amount="eighth" and price.price < 60 
    
    group by product.name
"""

df = pd.read_sql(q, conn)
df.sort_values('avg(price.price)', ascending=False)
```
